# First we import a pre trained model from Transformers.

#### We use the Tokenizer to convert the texts to sequence of numbers that refer to these words and the model variable is assigned to the pretrained model.

```python
from transformers import AutoTokenizer,AutoModelForSeq2SeqLM
import sentencepiece

tokenizer=AutoTokenizer.from_pretrained('Helsinki-NLP/opus-mt-en-hi',use_fast=False)
model= AutoModelForSeq2SeqLM.from_pretrained('Helsinki-NLP/opus-mt-en-hi')
```

#### The model takes in the tokenized words, we set the padding to true so that it aligns to the padding of the data that is originally used for training the model.

```python
src_text =['I love math']

translated = model.generate(**tokenizer(src_text, return_tensors="pt",padding=True))
[tokenizer.decode(t, skip_special_tokens=True) for t in translated]
```

```python
def translator(from_text):
  translation = model.generate(**tokenizer(from_text, return_tensors="pt",padding=True))
  result=tokenizer.decode(translation[0],skip_special_tokens=True)
  return result
```

#### The following code uses the gradio library to create the GUI.C

```python
#gradio app
import gradio as gr
gr.Interface(fn=translator,
             inputs=gr.inputs.Textbox(lines=2,placeholder='Text to translate')
             ,outputs='text').launch(share=True)
```
